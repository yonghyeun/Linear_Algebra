# Ax = C 를 만족하는 X 의 개수 찾기 

1. Ax = C 를 만족하는 해의 개수가 하나만 존재하기 위한 조건

    * Ax = C를 만족하는 해의 개수가 하나만 존재하려면 A 행렬의 열 벡터들이 서로 독립적이여야 한다. 이는 열 벡터들이 기저 벡터로만 이루어져 있는 경우를 포함한다. 독립적인 열 벡터의 개수가 해당 행렬의 열 개수와 같을 필요는 없다. 독립적인 열 벡터의 개수가 행렬의 랭크와 같을 때, Ax = C의 해가 하나만 존재한다.
        
        * 역행렬 관점
        
        * 1. 기저 벡터로만 이뤄진 행렬 A 는 벡터 공간에서 2차원 이상을 채울 수 있기에 역행렬이 존재한다.
        * 2. Ax = C 에서 양 변에 A의 역행렬을 곱해주면 X = A^-1 * C 벡터를 갖는다.

        * 기하학적 관점

        * 1. A의 벡터들은 차원의 개수만큼의 기저 벡터가 존재하여 모든 차원을 span 할 수 있다.
        * 2. Ax = C 에서 C 는 A 의 선형 결합으로 표현 가능한 벡터이다.

### Ax = C 란 결국 A 의 행벡터와 X 의 열벡터 간 선형 결합한 값으로 새로운 벡터를 생성한다는 것이다.

* 여기서 Ax = C 를 만족하는 X 의 벡터가 존재하는지, 존재하지 않는지, 무한하게 많은지에 대해서는 어떻게 알 수 있을까 ? 

# 2. Ax = C 에서 벡터 X 가 존재하기 위한 조건 

1. 행렬 A 와 C 의 차원이 일치해야 한다. 2차원 내 벡터 공간의 벡터들의 선형 결합으로는 3차원의 벡터를 표현 할 수 없다.
2. A 와 C 의 랭크가 동일해야 한다.
3. A의 열벡터들이 선형 독립적이여야 한다. 만약 선형 종속적인 경우 Ax = C 를 만족하는 해가 없거나 무수히 많을 수 있다.

* 1번은 직관적으로 이해가 된다. 
* 하지만 2번과 3번에 있는 선형 독립이라는 말에 대해서 잘 이해가 가지 않는다. 그리고 선형 종속적인 경우 Ax = C 를 만족하는 해가 없거나 무수히 많을 수 있는 경우는 뭐를 의미할까 ? 

# Rank 란

* Rank란 행렬 A 를 기약행 사다리꼴 형태로 변환 하였을 때 Pivot column 의 갯수를 의미한다. 각 Pivot column 의 갯수는 해당 행렬을 구성하는 선형 독립적인 행과 열의 최대 개수와 동일하다 

# Full Rank

### 해당 행렬의 벡터들이 모두 독립적인 상태 

```python
# Full Rank 행렬 A 생성

A = np.array([[1,2],[0,1]])

row_vectors = A[0,:], A[1,:]
col_vectors = A[:,0], A[:,1]

print('A 행렬의 형태')
display(A)

print('A의 랭크는')
print(np.linalg.matrix_rank(A))

fig,ax = plt.subplots(ncols = 2,figsize = (15,5))

ax[0].set_xlim([-3,3])
ax[0].set_ylim([-3,3])

ax[0].axvline(0, c = 'k')
ax[0].axhline(0, c = 'k')

ax[0].quiver(0,0,row_vectors[0][0], row_vectors[0][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'blue')
ax[0].quiver(0,0,row_vectors[1][0], row_vectors[1][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'red')

ax[0].text(row_vectors[0][0], row_vectors[0][0], 'V1 \n [1,2]')
ax[0].text(row_vectors[1][0] - 1, row_vectors[1][0] + 1, 'V2 \n [2,0]')

ax[0].set_title('Row vectors')


ax[1].set_xlim([-3,3])
ax[1].set_ylim([-3,3])

ax[1].axvline(0, c = 'k')
ax[1].axhline(0, c = 'k')

ax[1].quiver(0,0,col_vectors[0][0], col_vectors[0][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'blue')
ax[1].quiver(0,0,col_vectors[1][0], col_vectors[1][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'red')

ax[1].text(col_vectors[0][0] , col_vectors[0][1], 'V1 \n [1,0]')
ax[1].text(col_vectors[1][0], col_vectors[1][1], 'V2 \n [2,1]')

ax[1].set_title('Col vectors')

fig.suptitle('A의 행벡터, 열 벡터 \n Rank : 2',y = 1.05) ;
```

* A 행렬의 벡터들은 서로 독립적인 관계로 어떤 벡터도 서로간의 선형 결합으로 표현 될 수 없다.
* 즉, 독립적인 벡터가 두 개 있으므로 해당 2차원 벡터 공간에서 어떠한 벡터든 서로간의 선형 결합으로 표현 가능하다. 

### 그 말은 즉 Ax = C 를 만족하는 벡터 X가 유일하게 하나 존재 한다는 뜻이다. 

# Rank Deficient

### 해당 행렬의 벡터들이 모두 독립적이지 않고 종속적인 벡터가 존재하는 경우 

```python
# Rank Deficient 행렬 B 생성 

B = np.array([[1,2],[1,2]])
print('B 행렬의 형태 ')
display(B)
print('B의 랭크는')

print(np.linalg.matrix_rank(B))

# Full Rank 행렬 A 생성

row_vectors = B[0,:], B[1,:]
col_vectors = B[:,0], B[:,1]

fig,ax = plt.subplots(ncols = 2,figsize = (15,5))

ax[0].set_xlim([-3,3])
ax[0].set_ylim([-3,3])

ax[0].axvline(0, c = 'k')
ax[0].axhline(0, c = 'k')

ax[0].quiver(0,0,row_vectors[0][0], row_vectors[0][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'blue')
ax[0].quiver(0,0,row_vectors[1][0], row_vectors[1][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'red')

ax[0].text(row_vectors[0][0], row_vectors[0][0], 'V1 \n [1,2]')
ax[0].text(row_vectors[1][0] - 1, row_vectors[1][0] + 1, 'V2 \n [1,2]')

ax[0].set_title('Row vectors')


ax[1].set_xlim([-3,3])
ax[1].set_ylim([-3,3])

ax[1].axvline(0, c = 'k')
ax[1].axhline(0, c = 'k')

ax[1].quiver(0,0,col_vectors[0][0], col_vectors[0][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'blue')
ax[1].quiver(0,0,col_vectors[1][0], col_vectors[1][1], angles = 'xy', scale = 1, scale_units = 'xy', color = 'red')

ax[1].text(col_vectors[0][0] , col_vectors[0][1], 'V1 \n [1,1]')
ax[1].text(col_vectors[1][0], col_vectors[1][1], 'V2 \n [2,2]')

ax[1].set_title('Col vectors')

fig.suptitle('B의 행벡터, 열 벡터 \n Rank :image.png 1',y = 1.05) ;
```

* B 행렬의 벡터들 중 하나는 종속적인 관계를 가지고 있다. 행 벡터들은 서로간의 벡터 값이 같으며, 열 벡터들은 V2 = V1 * 2 로 표현 가능하다.
* 즉 B 행렬의 벡터는 독립적인 벡터가 하나밖에 없어 Rank 값을 하나만 갖게 된다. 독립적인 벡터가 하나밖에 존재하지 않는다는 것이다.

# 해가 무수히 많을 경우에 대한 시각화

```python
A = np.array([[2,1],[0,0]])

C = [0,1,2,3,4,5]

v1 = A[:,0]
v2 = A[:,1]

fig,ax = plt.subplots(figsize = (10,10))

ax.axhline(0, color = 'k')
ax.axvline(0, color = 'k')

ax.set_xlim([-3,3], )
ax.set_ylim([-3,3], )


ax.quiver(0,0,v1[0],v1[1],angles = 'xy', scale_units = 'xy',color = 'blue',scale = 1, label = 'V1')
ax.quiver(0,0,v2[0],v2[1],angles = 'xy', scale_units = 'xy',color = 'red',scale = 1, label = 'V2')

X = np.linspace(-6,6,100)

for c in C:
    
    Y = 2 * -X + c
    
    ax.plot(X,Y, label = f'Solution space (C = [{c},0])', lw = 1, ls = '--')

    
ax.legend(loc = 'upper left')
ax.set_title('Ax = C 를 만족하는 X의 solution space', fontsize = 20)
```

```python
# 예시 Ax = 5 를 만족하는 x 의 solution space 와 예시 벡터들

C = 5

X = np.linspace(-5,5,100)
Y = 2 * -X + C

null_Y = 2 * -X

fig, ax = plt.subplots(figsize = (10,10))

ax.axhline(0, color = 'k')
ax.axvline(0, color = 'k')

ax.set_xlim([-5,5], )
ax.set_ylim([-5,5], )

ax.plot(X,Y, c = 'green', ls = '--', lw = 0.5, label = 'Ax = 5') # Ax = 5 를 만족하는 solution space
ax.plot(X,null_Y, c = 'red', ls = '--', lw = '0.5', label = 'Ax = 0') # Ax = 0 을 만족하는 null space
for i in range(5):
    
    solution_idx = np.random.choice(range(65,90))
    solution_vector = np.array([X[solution_idx], Y[solution_idx]])
    
    null_idx = np.random.choice(range(40,60))
    null_vector = np.array([X[null_idx], null_Y[null_idx]])
    
    ax.quiver(0,0,solution_vector[0], solution_vector[1], scale = 1, scale_units = 'xy', color = 'green')
    ax.text(solution_vector[0], solution_vector[1], f'solution vector {i}')
    
    ax.quiver(0,0,null_vector[0], null_vector[1], scale = 1, scale_units = 'xy', color = 'red')
    ax.text(null_vector[0] - 1.5, null_vector[1], f'null vector {i}')

    
ax.legend()
    
ax.set_title('Ax = [0,5] 를 만족하는 Solution vector 와 Solution Space' , fontsize = 20)
```

```python
# 예시 Ax = 5 를 만족하는 x 의 solution space 와 예시 벡터들

C = 5

X = np.linspace(-5,5,100)
Y = 2 * -X + C

null_Y = 2 * -X

fig, ax = plt.subplots(figsize = (10,10))

ax.axhline(0, color = 'k')
ax.axvline(0, color = 'k')

ax.set_xlim([-5,5], )
ax.set_ylim([-5,5], )

ax.plot(X,Y, c = 'green', ls = '--', lw = 0.5, label = 'Ax = 5의 Solution Space') # Ax = 5 를 만족하는 solution space
ax.plot(X,null_Y, c = 'red', ls = '--', lw = '0.5', label = 'Ax = 5의 Solution Space') # Ax = 0 을 만족하는 null space


# Free variable 값이 어떤 임의의 값을 가질 때의 벡터 생성 

solution_idx = np.random.choice(range(65,90))
solution_vector = np.array([X[solution_idx], Y[solution_idx]])

ax.quiver(0,0,solution_vector[0], solution_vector[1], scale = 1, scale_units = 'xy', color = 'green', label = 'Particulary vector')
ax.text(solution_vector[0] + 0.5, solution_vector[1], 'Particulary vector')


# null space 위에 있는 벡터 하나 선정

null_idx = np.random.choice(range(40,60))
null_vector = np.array([X[null_idx], null_Y[null_idx]])
    
ax.quiver(0,0,null_vector[0], null_vector[1], scale = 1, scale_units = 'xy', color = 'red', label = 'null vector')
ax.text(null_vector[0] - 1.5, null_vector[1], 'Null vector')

# Null space 위의 벡터와 Particulary Vector 의 선형 결합하여 Completed Solution 생성

completed_solution = solution_vector + null_vector
ax.quiver(0,0,completed_solution[0], completed_solution[1], scale = 1, scale_units = 'xy', color = 'Blue', label = 'Completed vector')
ax.text(completed_solution[0], completed_solution[1], 'Null Vector + Particulary Vector \n (Completed Vector)')


    
ax.legend()
    
ax.set_title('Completed Solution 을 구하기 위한 \n Null Vector 와 Particulary Vector 의 선형 결합' , fontsize = 20)
```
