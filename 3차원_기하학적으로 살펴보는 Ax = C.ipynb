# Ax = 0 을 만족하는 벡터 X 찾기 

### 1. X의 해가 유한한 경우 

```python
A = np.array([[1,3,1],[2,2,1], [2,3,1]])

C = np.array([4,-1,3])

print('A 행렬의 형태는')
print(A)
```

* A 행렬은 3차원 공간 내의 3개의 벡터를 나타낸다 

```python
v1 = A[:,0]
v2 = A[:,1]
v3 = A[:,2]


fig,ax = plt.subplots(figsize = (10,10),
                        subplot_kw = {'projection' : '3d'})

ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
ax.set_zlim([-5,5])

ax.set_xlabel('x', fontsize = 20)
ax.set_ylabel('y', fontsize = 20)
ax.set_zlabel('z', fontsize = 20)

# x축을 나타내는 직선
ax.plot([-5, 5], [0, 0], [0, 0], color='k', lw = 0.5)
# y축을 나타내는 직선
ax.plot([0, 0], [-5, 5], [0, 0], color='k', lw = 0.5)
# z축을 나타내는 직선
ax.plot([0, 0], [0, 0], [-5, 5], color='k', lw = 0.5)

colors = ['orange','blue','green']
names = ['v1','v2','v3']
for i,vector in enumerate([v1,v2,v3]):
    
    name = str(vector)
    
    ax.quiver(0,0,0,vector[0],vector[1],vector[2],
                color = colors[i], alpha = 0.3, arrow_length_ratio = 0.1)
    ax.text(vector[0],vector[1],vector[2], f'{names[i]} \n {name}', color = colors[i])

    
ax.set_title('3차원 벡터 공간에서 A 행렬의 벡터', fontsize = 20)
```

### 벡터 공간에서 행렬과 벡터의 곱은 선형 결합

* 그렇다면 최종적으로 생성되는 [4,2,5] 벡터는 A 행렬의 벡터들을 적당히 선형 결합해서 생성된 벡터임을 알 수 있다.

* 어떤 벡터가 행렬 A가 생성 할 수 있는 벡터 공간 내에 있다면 해는 존재한다.

```python
_,R = np.linalg.qr(A)

v1 = R[:,0]
v2 = R[:,1]
v3 = R[:,2]

fig,ax = plt.subplots(figsize = (10,10),
                        subplot_kw = {'projection' : '3d'})

ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
ax.set_zlim([-5,5])

ax.set_xlabel('x', fontsize = 20)
ax.set_ylabel('y', fontsize = 20)
ax.set_zlabel('z', fontsize = 20)

# x축을 나타내는 직선
ax.plot([-5, 5], [0, 0], [0, 0], color='k', lw = 0.5)
# y축을 나타내는 직선
ax.plot([0, 0], [-5, 5], [0, 0], color='k', lw = 0.5)
# z축을 나타내는 직선
ax.plot([0, 0], [0, 0], [-5, 5], color='k', lw = 0.5)

colors = ['orange','blue','green']
names = ['v1','v2','v3']
for i,vector in enumerate([v1,v2,v3]):
    
    
    ax.quiver(0,0,0,vector[0],vector[1],vector[2],
                color = colors[i], arrow_length_ratio = 0.1)
    ax.text(vector[0],vector[1],vector[2], names[i], color = colors[i])
    
ax.view_init(5,10)
    
ax.set_title('벡터 공간 내 A 행렬의 벡터들의 방향 \n V1 : X축 , V2 : Y축 , V3 : Z축', fontsize = 20)
```

* A행렬의 벡터들은 모두 독립적으로 모든 차원을 향해 있기 때문에 3차원 내의 모든 벡터를 Span 할 수 있게 된다. 
* v1,v2,v3 벡터들은 3 차원 벡터 공간의 기저벡터들을 얼만큼을 스칼라배해서 생긴 벡터임을 알 수 있다.
* 그 말은 v1,v2,v3 벡터들을 다시 스칼라만큼을 나눠준다면 기저 단위벡터 형태로 돌아 갈 수 있음을 의미한다 


* Ax = C 에서 A 행렬과 C벡터를 첨가행렬 형태인 A|C 형태로 변환 후 A행렬을 RREF 형태로 만들어주는 과정은 양 변에 A를 역행렬로 만들어주는 순열 행렬을 계속 곱해주는 행위와 같다.

* 즉 Ax = C 에서 양 변에 A의 역행렬을 곱해줌으로서 X = A의 역행렬 * C 벡터 라는 값을 가지게 되며 X 벡터를 구할 수 있게 된다 

```python
A_inverse = np.linalg.inv(A)

I = A_inverse @ A

i1 = I[:,0]
i2 = I[:,1]
i3 = I[:,2]

solution = A_inverse @ C

fig,ax = plt.subplots(figsize = (10,10),
                        subplot_kw = {'projection' : '3d'})

ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
ax.set_zlim([-5,5])

ax.set_xlabel('x', fontsize = 20)
ax.set_ylabel('y', fontsize = 20)
ax.set_zlabel('z', fontsize = 20)

# x축을 나타내는 직선
ax.plot([-5, 5], [0, 0], [0, 0], color='k', lw = 0.5)
# y축을 나타내는 직선
ax.plot([0, 0], [-5, 5], [0, 0], color='k', lw = 0.5)
# z축을 나타내는 직선
ax.plot([0, 0], [0, 0], [-5, 5], color='k', lw = 0.5)

colors = ['orange','blue','green']
names = ['i1','i2','i3']
for i,vector in enumerate([i1,i2,i3]):
    
    
    ax.quiver(0,0,0,vector[0],vector[1],vector[2],
                color = colors[i], arrow_length_ratio = 0.1)
    ax.text(vector[0],vector[1],vector[2], names[i], color = colors[i])

ax.quiver(0,0,0,solution[0],solution[1],solution[2], arrow_length_ratio = 0.1, color = 'red')
ax.text(solution[0],solution[1],solution[2], f'X vector \n {solution}', color = 'red')

ax.set_title('RREF 형태로 변환하며 구한 X vector', fontsize = 20)
```

# Ax = C 를 만족하는 벡터는 무한하게 많은데 Ax = C 를 만족하는 벡터는 다음과 같다. 

### X completed = X_0 (Ax = 0 을 만족하는 벡터) + X_particulary vector (Ax = C 를 만족하는 벡터 중 일부)
