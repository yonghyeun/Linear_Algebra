```python
import matplotlib.pyplot as plt 
import numpy as np

plt.style.use('seaborn-whitegrid')
plt.rc('font', family = 'malgun gothic')
plt.rcParams['axes.unicode_minus'] = False
```

# 두 벡터의 내적은 왜 한 벡터를 다른 벡터에 정사영 한 후 곱한 값과 같을까 ? 

##### 기하학적으로 이해하는 선형 대수 

url : https://www.youtube.com/watch?v=ZH79kAgC3I4&t=714s


* 두 벡터의 내적은 [2 1] @ [X Y]^T = 2*X + 1 * Y 원소 곱의 합으로 나타낼 수 있음 
* 또 두 벡터의 내적은 [X Y] 벡터를 [2 1] 벡터 위에 정사영 한 후 두 벡터의 놈의 곱의 합으로 나타낼 수 있음 
    * 왜 그런지에 대해 그래프를 그리며 나타내봄 

```python
v1 = [2,1] # 행 벡터 [2,1] 이 존재 

# v1 @ v2 = C 일 때 v2 벡터 ([X Y])가 가질 수 있는 영역을 시각화 해볼 수 있음 


print('v1 : [2,1]')
print('v2 : [x,y]')
print('v1 @ v2 = C 일 때 v2의 원소인 x,y 가 가질 수 있는 순서 쌍 공간  (@ 은 dot product)')
```

```python
C = [-1,0,1,2,3,4] # 두 벡터가 내적한 값이 다음과 같을 때 
X = np.linspace(-5,5,100) # X 값과
fig,ax = plt.subplots(figsize = (7,7))

ax.axvline(0, c = 'k')
ax.axhline(0, c = 'k')

colors = ['red','orange','green','purple','salmon','brown']

for i,c in enumerate(C):
    
    Y = (c - v1[0] * X) / v1[1] # Y 값이 가질 수 있는 값들
    
    ax.plot(X,Y, ls = '--', c = colors[i], label = f'C = {c}') # [X Y] 벡터가 가질 수 있는 영역
    ax.legend(frameon = True, ncol = 2)
    
ax.scatter(v1[0], v1[1], c = 'blue') # 행 벡터 좌표
ax.quiver(0,0,v1[0],v1[1], scale = 8, color = 'blue') # 행 벡터 
ax.text(2,1.1,' 행벡터 V1 (2,1)', fontsize = 15)

ax.set_xlim([-3,5])
ax.set_ylim([-3,5])
ax.set_xticks(np.arange(-3,6,1))
ax.set_yticks(np.arange(-3,6,1))

ax.set_title('[2 1] @ [x y]^T = C 일 때 \n C 의 값에 따라 x,y 가 가질 수 있는 벡터의 영역',
                fontsize = 15);
```

### 1. 내적 값 C 에 따라 생기는 열 벡터의 공간 직선 그래프는 행벡터 V1 과 수직으로 만난다.
### 2. 내적 값 C 는 행벡터가 열벡터의 등고선을 지나가는 개수와 같다 

```python
X = np.linspace(-5,5,100)


fig,ax = plt.subplots(figsize = (7,7))

ax.axvline(0, c = 'k')
ax.axhline(0, c = 'k')

c = 4
    
Y = (c - v1[0] * X) / v1[1]

ax.plot(X,Y, ls = '--', label = f'C = {c}', c = 'brown')
ax.legend(frameon = True)
    
ax.scatter(v1[0], v1[1])

    
ax.quiver(0,0,X[60], Y[60], color = 'brown', scale = 8)
ax.text(X[60], Y[60], '벡터 V2', fontsize = 15)

ax.quiver(0,0,v1[0],v1[1], scale = 8, color = 'blue', alpha = 0.3)

ax.plot([0,1.6],[0,0.8], c = 'red') # C = 4 일 때 삼각형의 높이이자 벡터 V2 를 V1 에 정사영 시킨 값 
ax.text(1.3,0.5,'삼각형의 높이', weight = 'bold')

ax.set_xlim([-3,5])
ax.set_ylim([-3,5])
ax.set_xticks(np.arange(-3,6,1))
ax.set_yticks(np.arange(-3,6,1))

ax.set_title('C = 4 일 때 벡터 [X Y] 들의 예시',
                fontsize = 15);
```

# 삼각형의 높이는 벡터 V2 를 행벡터 V1 위에 정사영 시킨 값과 같음
