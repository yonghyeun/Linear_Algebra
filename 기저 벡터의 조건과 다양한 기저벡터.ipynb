# 3차원 벡터 공간의 기저 벡터 

* 기저 벡터의 조건 

1. 선형 독립 (Linear Independence): 기저벡터들은 서로 선형 독립이어야 합니다. 즉, 어떤 벡터도 다른 기저벡터들의 선형 조합으로 표현될 수 없어야 합니다.

2. 공간을 생성 (Spanning): 기저벡터들의 선형 조합으로 어떤 벡터도 생성할 수 있어야 합니다. 즉, 기저벡터들의 선형 조합으로 모든 벡터를 표현할 수 있어야 합니다.

3. 최소 개수 (Minimal): 기저벡터의 개수는 해당 공간을 생성하는 데 필요한 최소한의 벡터 개수여야 합니다. 즉, 불필요한 벡터를 제외하고도 해당 공간을 표현할 수 있어야 합니다.

4. 일차독립 (Linearly Independent): 기저벡터들은 선형 독립이어야 하며, 어떤 기저벡터를 제외하고도 여전히 선형 독립이어야 합니다. 기저벡터 중 하나를 제거하면 해당 공간을 생성하는 능력이 상실되면 안 됩니다.

```python
dim = 3 
basis_vector = np.eye(dim, dtype = float)

i_1 = basis_vector[:,0]
i_2 = basis_vector[:,1]
i_3 = basis_vector[:,2]

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(projection = '3d')

ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
ax.set_zlim([-3,3])

# x축을 나타내는 직선
ax.plot([-3, 3], [0, 0], [0, 0], color='k', lw = 0.3)
# y축을 나타내는 직선
ax.plot([0, 0], [-3, 3], [0, 0], color='k', lw = 0.3)
# z축을 나타내는 직선
ax.plot([0, 0], [0, 0], [-3, 3], color='k', lw = 0.3)

colors = ['red','green','blue']

for i in range(3):
    ax.quiver(0,0,0,basis_vector[:,i][0],basis_vector[:,i][1],basis_vector[:,i][2],
                color = colors[i])
    ax.text(basis_vector[:,i][0],basis_vector[:,i][1],basis_vector[:,i][2],
            basis_vector[:,i])
    
ax.set_title('3 차원의 기저 벡터', fontsize = 20)
```

```python
tril_matrix = np.tril(np.ones((3,3)))
triu_matrix = np.triu(np.ones((3,3)))


print('하삼각행렬의 형태')
print(tril_matrix)
print('\n\n')
print('상삼각행렬의 형태')
print(triu_matrix)
```
